```python
#imports
import requests
import selenium
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pyautogui as pag
import re
import pandas as pd
```

```python
url=('https://webp.caixa.gov.br/portal/consulta/proponente_nv.asp')
driver = webdriver.Chrome('/home/sol/notebooks/web_scraper_python/chromedriver')
cpf_list = ['20609361015',
'00180674099',
'00072585005',
'40772721068',
'31664636072',
'29594650087',
'36656755020',
'00592895033',
'26191644000',
'33324492068',
'45589526000',
'23954035049',
'37882651087',
'21564230082',
'32091303020',
'81694385000',
'47882840068',
'43242170091',
'82589844034',
'33915407020']

results = []
for cpf in cpf_list:
    driver.get(url)
    driver.maximize_window()
    driver.find_element_by_id('txtCPF').send_keys(cpf)
    driver.find_element_by_id('btnBuscar').click()
    time.sleep(6)
    container_elements = driver.find_elements_by_class_name('container')
    consulta = container_elements[2].text
    pattern = re.compile(r'ESTE.+S.')
    pre_result = re.findall(r'ESTE.+S.', str(consulta))
    results.append(pre_result)
    time.sleep(2)
```

```python
df = pd.DataFrame(list(zip(cpf_list, results)), columns = ['CPF', 'RESULTADO'])
df




```
